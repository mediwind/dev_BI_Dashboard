```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

tables = pd.read_html(url)

sp500_table = tables[0]
sp500_table
```

```python
sp500_tickers = sp500_table['Symbol']
sp500_tickers
```

```python
import yfinance as yf

def get_sp500_market_cap(sp500_tickers):
    market_caps = dict()
    
    for ticker in sp500_tickers:
        try:
            info = yf.Ticker(ticker).info
            symbol = info['symbol']
            market_cap = info['marketCap']
            market_caps[symbol] = market_cap
        except Exception as e:
            print(f'Error fetching data for {ticker}: {e}')
    
    return market_caps

sp500_market_caps = get_sp500_market_cap(sp500_tickers)

for symbol, market_cap in sp500_market_caps:
    print(f'{symbol}: {market_cap}')
```

```python
sp500_market_caps['BRK.B'] = 787310000000
sp500_market_caps['BF.B'] = 26500000000

print(len(sp500_market_caps))

for symbol, market_cap in sp500_market_caps.items():
    print(f'{symbol}: {market_cap}')
```

```python
sp500_df = pd.DataFrame(list(sp500_market_caps.items()), columns=['Symbol', 'Market_Capital'])

merged_df = pd.merge(sp500_table, sp500_df, on='Symbol')

merged_df
```

```python
merged_df = merged_df.sort_values('Ticker')
merged_df.to_csv('Details_sp500.csv')
```
