```python
import pandas as pd

res = pd.read_csv('Details.csv')

res
```

```python
gics_sector = res['GICS Sector'].unique()
gics_sub_industry = res['GICS Sub-Industry'].unique()

gics_sector
gics_sub_industry
```

```python
from collections import defaultdict

dict_for_sector = defaultdict(list)
dict_for_industry = defaultdict(list)

for gs in gics_sector:
    dict_for_sector[gs].append(res[res['GICS Sector'] == gs][['Company', 'Ticker']])

for si in gics_sub_industry:
    dict_for_industry[si].append(res[res['GICS Sub-Industry'] == si][['Company', 'Ticker']])
```

```python
dict_for_sector
dict_for_industry
```

```python
for k, v in dict_for_industry.items():
    print('---------------', k, '------------------')
    print()
    print(k, v)
    print('\n\n')
```
