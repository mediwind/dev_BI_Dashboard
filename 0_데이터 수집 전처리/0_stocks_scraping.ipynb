### AAPL로 테스트

### 나스닥 100 티커 가져오기

```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/NASDAQ-100'

tables = pd.read_html(url)

nasdaq_100_table = tables[4]

tickers = nasdaq_100_table['Ticker']

print(tickers)
```

### 티커들을 이용해 101개의 나스닥 csv 파일 생성

```python
for ticker in tickers:
    data = yf.download(f'{ticker}', start='2000-01-01', end='2023-11-30')

    close_data = data['Close']

    close_data.to_csv(f'stocks_price/{ticker}_close.csv')
```

### (비권장) 조건 없는 나스닥 100 파일들 Concat

```python
import pandas as pd
import os

directory = 'stocks_price'

files = [f for f in os.listdir(directory) if f.endswith('.csv')]

dfs = []

for file in files:
    df = pd.read_csv(os.path.join(directory, file))
    df = df.rename(columns={'Close': os.path.splitext(file)[0]})
    dfs.append(df)

merged_df = pd.concat(dfs, axis=1).fillna('')
merged_df.to_csv('merged.csv', index=False)
```

### (권장) AAPL을 시작으로 나머지 나스닥 주식들을 Date 컬럼 기준으로 merge

```python
import pandas as pd

directory = 'stocks_price'

df1 = pd.read_csv(f'{directory}/AAPL_close.csv')
flag = 0

for ticker in tickers[1:]:
    if flag:
        df1 = pd.read_csv(f'{directory}/merged.csv')
    
    df2 = pd.read_csv(f'{directory}/{ticker}_close.csv')
    df2 = df2.rename(columns={'Close': f'{ticker}'})

    merged_df = df1.merge(df2, on='Date', how='left')
    merged_df.to_csv(f'{directory}/merged.csv', index=False)
    
    flag = 1
```
