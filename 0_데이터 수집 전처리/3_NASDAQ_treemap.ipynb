```python
import yfinance as yf

def get_nasdaq100_market_cap(nasdaq100_tickers):
    market_caps = {}

    for ticker in nasdaq100_tickers:
        try:
            info = yf.Ticker(ticker).info
            symbol = info['symbol']
            market_cap = info['marketCap']
            market_caps[symbol] = market_cap
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")

    return market_caps

nasdaq100_market_caps = get_nasdaq100_market_cap(nasdaq100_tickers)

for symbol, market_cap in nasdaq100_market_caps.items():
    print(f"{symbol}: {market_cap}")
```

```python
import pandas as pd

nasdaq_df = pd.DataFrame(list(nasdaq100_market_caps.items()), columns=['Ticker', 'Market_Capital'])

merged_df = pd.merge(res, nasdaq_df, on='Ticker')
merged_df = merged_df.sort_values('Ticker')
```
