```python
%%sql

CREATE TABLE raw_data.unemployment (
    "date" VARCHAR(50),
    "value" NUMERIC(3, 1)
);
```

```python
%%sql

ALTER TABLE raw_data.unemployment
ADD COLUMN formatted_date date;

-- 'formatted_date' 값을 계산하여 업데이트
UPDATE raw_data.unemployment
SET formatted_date = TO_DATE(date, 'MM/DD/YYYY');
```

```python
%%sql

SELECT * FROM raw_data.unemployment;
```

```python
%%sql

CREATE TABLE analytics.unemployment_price_summary AS
SELECT B.*, A.value as unemployment
FROM raw_data.unemployment A
JOIN raw_data.prices B ON TO_CHAR(A.formatted_date, 'YYYY-MM') = TO_CHAR(B.date, 'YYYY-MM');
```

```python
%%sql

select * from analytics.unemployment_price_summary limit 10;
```
